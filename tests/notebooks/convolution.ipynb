```python
import numpy as np
from matplotlib import pyplot as plt
from nanopyx.core.transform._le_convolution import Convolution

prg = Convolution()
```

```python
%matplotlib widget
run_types = prg._run_types
fig,axs = plt.subplots(1,len(run_types))

img_dims = 1000
img = np.ones((img_dims, img_dims)).astype(np.float32)

kernel_dims = 7
kernel = np.zeros((kernel_dims, kernel_dims)).astype(np.float32)
kernel[int(kernel_dims/2)-1:int(kernel_dims/2)+1, :] = 1
kernel[:, int(kernel_dims/2)-1:int(kernel_dims/2)+1] = 1

for i, run_type in enumerate(run_types):
    for n in range(3):
        output = prg.run(img, kernel, run_type=run_type)
    axs[i].imshow(output)
    axs[i].set_title(run_type)

plt.show()
```
