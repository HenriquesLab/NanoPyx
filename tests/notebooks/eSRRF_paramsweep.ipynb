```python
import numpy as np
from nanopyx.core.transform._le_esrrf import eSRRF
```

```python
imgs = [np.random.random((100, i, i)) for i in [100, 200, 400, 800]]
```

```python
runner = eSRRF()
for img in imgs:
    for mag in range(1, 5):
        for rad in range(1, 5):
            for sensitivity in range(1, 5):
                for i in range(3):
                    print(f"img: {img.shape}, mag: {mag}, rad: {rad*0.4}, sensitivity: {sensitivity*0.4}")
                    runner.benchmark(img, magnification=mag, sensitivity=sensitivity*0.4, radius=rad*0.4)
```
