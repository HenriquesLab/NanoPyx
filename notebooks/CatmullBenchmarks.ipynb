```python
import pandas as pd
fld = "Colab"
data_path = "CatmullRomBenchmarks/BenchmarksCatmull" + fld + ".csv"

df = pd.read_csv(data_path, index_col=0)
df
```

```python
import seaborn as sns
from matplotlib import pyplot as plt
from matplotlib.colors import LogNorm

sns.color_palette("Spectral", as_cmap=True)

plt.figure(figsize=(6,5))
ax = sns.heatmap(df, annot=True, cmap="Spectral",
                 norm=LogNorm(),
                 fmt=".2f")
plt.title("Run time ratio between OpenCL and other run types")
plt.ylabel("Run type")
plt.xlabel("Input shape (Frames x Height x Width)")
plt.xticks(rotation=45)
plt.savefig("/Users/bsaraiva/Code/NanoPyx/ratio_catmull_benchmarks_" + fld + ".pdf", dpi=300, bbox_inches = 'tight')
plt.show()
```
